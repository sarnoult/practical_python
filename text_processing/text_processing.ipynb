```python
import gutenberg as gb
```

```python
url = "https://www.gutenberg.org/ebooks/1342.txt.utf-8"
```

It takes time to download the book every time. In the new version of `read_text`, the text is saved to a local cache. 
After modifying `read_text` in `gutenberg.py`, run the following cell twice. You should see that the execution time drops from about 2.5 seconds to 0.0s: the first time because the text is downloaded, the second time because it is read from the local 
cache file.

```python
# the root directory for the notebook is 'text_processing' (this is the directory where
# the notebook is located), so our save_dir should be `data`, and not the default 
# `text_processing/data`.
txt = gb.read_text(url, save_dir='data')    
```

```python
lines = txt.splitlines()
```
